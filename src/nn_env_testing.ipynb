```python
import torch

if torch.backends.mps.is_available():
    device = torch.device("mps")
    print(f"Using MPS: {device}")
else:
    device = torch.device("cpu")
    print(f"Using CPU: {device}")


if device.type == "mps":
    x = torch.ones(5, device=device)
    print(f"Test Tensor on MPS: {x.device}")
```
